```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import time

all_list = []

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=chrome_options)

url = "https://www.imdb.com/title/tt23849204/reviews/?ref_=ttrt_sa_3"
driver.get(url)

while True: # infinite loop (untill breaks)
    try:
        load_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CLASS_NAME, 'ipl-load-more__button'))
        )
        load_button.click() # if button found , then it will be clicked
        time.sleep(2)  # Waiting for reviews to load
    except Exception as e:
        print("No more reviews to load. Error:", str(e))
        break

response = driver.page_source
driver.quit()

soup = BeautifulSoup(response, 'html.parser')
result = soup.find_all(class_='review-container')
```

```python

# this is for another movie
import requests
from bs4 import BeautifulSoup
url = "https://www.imdb.com/title/tt11821912/reviews/?ref_=tt_ov_urv"
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:123.0) Gecko/20100101 Firefox/123.0",
    "Accept": "text/html",
    "Connection": "keep-alive"
}
response1 = requests.get(url, headers=headers)
response_result1 = BeautifulSoup(response1.content, 'html.parser')
result1 = response_result1.find_all(class_='review-container')
# extracted same result and conncated them in final_result
final_result=result+result1
```

```python
all_list = []

# Collecting usernames and other information
def collect_usernames(item):
  username = item.a.get_text()
  return username

# Collecting Text data (Review Given)
def review_text(item):
  review_text = item.find(class_='text').get_text()
  return review_text

# Rating given on scale of 10
def rating_given(item):
  rating_element = item.find(class_='rating-other-user-rating')
  if rating_element:
        rating = rating_element.get_text().strip()  # Extract text and strip whitespace
        return rating
  else:
        return "N/A"  # Default value if no rating is found

# loop1
for item in final_result:
    user_name=collect_usernames(item.find(class_='display-name-link'))
    review=review_text(item)
    rating=rating_given(item)
    # appendding user_name,review,rating
    all_list.append({
        'username': user_name,
        'review': review,
        'rating': rating
    })
```

```python
for entry in all_list:
    print(f"Rating: {entry['rating']}")
    print(f"Username: {entry['username']}")
    print(f"Review: {entry['review']}")
    print()
```

```python
import csv
with open('reviews_1.csv', mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=['username', 'review', 'rating'])
    writer.writeheader()
    writer.writerows(all_list)
    print("CSV created.")
```

# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
# Write code for each of the sub parts with proper comments.
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
import pandas as pd
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stop_words = set(stopwords.words('english'))
```

```python

def text_cleaner(text):
  review_text = re.sub(r'[^\w\s]', '', text)
  review_text = re.sub(r'\d+', '', review_text)
  review_text = review_text.lower()
  filtered_words = [word for word in review_text.split() if word.lower() not in stop_words]
  stemmed_words = [stemmer.stem(word) for word in filtered_words]
  review_lemma = [lemmatizer.lemmatize(word) for word in stemmed_words]
  return ' '.join(review_lemma)
# The above code is taken from
cleaned_text = []
for item in all_list:
  cleaned_text.append(text_cleaner(item['review']))
```

```python
print(all_list[0]['review'])
print(cleaned_text[0])
```

```python
for i, item in enumerate(all_list):
    item['cleaned_review'] = cleaned_text[i]

with open('reviews_cleaned.csv', mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=['rating','review','username','cleaned_review'])
    writer.writeheader()
    writer.writerows(all_list)
    print("CSV file created.")
```

```python
# Your code here
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

```python
cleaned_text1 = pd.read_csv('reviews_cleaned.csv')
cleaned_review=cleaned_text1['cleaned_review']
from collections import Counter
pos_counts=Counter({'Nouns':0, 'Verbs':0, 'Adjectives':0,'Adverbs':0})
def pos_taggs(item):
  doc = nlp(item)
  for tockens in doc:
    # print(tockens.text, tockens.pos_)
    match tockens.pos_:
      case 'NOUN':
        pos_counts['Nouns']+=1
      case 'VERB':
        pos_counts['Verbs']+=1
      case 'ADJ':
        pos_counts['Adjectives']+=1
      case 'ADV':
        pos_counts['Adverbs']+=1

counts_review = cleaned_review.apply(pos_taggs)
print("\nPOS Counts:")
print(f"Total Nouns: {pos_counts['Nouns']}")
print(f"Total Verbs: {pos_counts['Verbs']}")
print(f"Total Adjectives: {pos_counts['Adjectives']}")
print(f"Total Adverbs: {pos_counts['Adverbs']}")
```
