# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
# Write code for each of the sub parts with proper comments.
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
import pandas as pd
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
def text_cleaner(text):
  review_text = re.sub(r'[^\w\s]', '', text)
  review_text = re.sub(r'\d+', '', review_text)
  review_text = review_text.lower()
  filtered_words = [word for word in review_text.split() if word.lower() not in stop_words]
  stemmed_words = [stemmer.stem(word) for word in filtered_words]
  review_lemma = [lemmatizer.lemmatize(word) for word in stemmed_words]
  return ' '.join(review_lemma)
# The above code is taken from
cleaned_text = []
for item in reviews:
  cleaned= text_cleaner(item)
  cleaned_text.append(cleaned)
```

```python
df1 =pd.DataFrame(cleaned_text, columns=['cleaned_review'])
df1.to_csv('reviews_cleaned1.csv', index=False)
```

```python
# Your code here
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

```python
cleaned_text1 = pd.read_csv('reviews_cleaned1.csv')
cleaned_review=cleaned_text1['cleaned_review']
from collections import Counter
pos_counts=Counter({'Nouns':0, 'Verbs':0, 'Adjectives':0,'Adverbs':0})
def pos_taggs(item):
  doc = nlp(item)
  for tockens in doc:
    # print(tockens.text, tockens.pos_)
    match tockens.pos_:
      case 'NOUN':
        pos_counts['Nouns']+=1
      case 'VERB':
        pos_counts['Verbs']+=1
      case 'ADJ':
        pos_counts['Adjectives']+=1
      case 'ADV':
        pos_counts['Adverbs']+=1

counts_review = cleaned_review.apply(pos_taggs)
print("\nPOS Counts:")
print(f"Total Nouns: {pos_counts['Nouns']}")
print(f"Total Verbs: {pos_counts['Verbs']}")
print(f"Total Adjectives: {pos_counts['Adjectives']}")
print(f"Total Adverbs: {pos_counts['Adverbs']}")
```
