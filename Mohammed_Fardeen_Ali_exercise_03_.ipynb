```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = "https://finviz.com/news.ashx?v=3"
all_items = []
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:123.0) Gecko/20100101 Firefox/123.0",
    "Accept": "text/html",
    "Connection": "keep-alive"
}
response = requests.get(url, headers=headers)
response_result = BeautifulSoup(response.content, 'html.parser')
print(response_result)
```

```python
news =[]
heading = response_result.findAll("a", class_="nn-tab-link")
for heading in heading:
  news.append(heading.get_text(strip=True))
news_sample= news[0:10]
print(news_sample)
```

```python
data_df = pd.DataFrame(news, columns=['News'])
pd.set_option('display.max_colwidth', None)
print(data_df.head(10))
```

```python
# text pre-processing

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
filtered_tockens_list = []

for sentence in news_sample:
    # Transforming the sentance
    sentence = sentence.lower()
    # Tockenizing sentance
    tokens = word_tokenize(sentence)
    # Removing stop words and special characters from the tockens
    filtered_tokens = [word for word in tokens if word not in stop_words]
    filtered_tockens_list.append(filtered_tokens)

    print(f"Original Sentence: {sentence}")
    print(f"Tokens without stop words: {filtered_tokens}")
    print()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd

# Example news headline
news1 = news_sample

# Initialize CountVectorizer
vect = CountVectorizer()
bow_matrix = vect.fit_transform(news1)
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vect.get_feature_names_out())

# intializeing tfidf
tfidf_vect = TfidfVectorizer()
tfidf_matrix = tfidf_vect.fit_transform(news1)

pd.DataFrame(
    # Converts the sparse matrix into a dense matrix (a more readable format).
    tfidf_matrix.todense(),
    columns=tfidf_vect.get_feature_names_out()
)

print("Vocabulary:", vect.vocabulary_)
print("Shape:", bow_df.shape)
print("Bag of Words Matrix:\n", bow_df)

print("TF-IDF Matrix:\n", tfidf_df)
```

```python
import spacy

# Load the spacy language model
nlp = spacy.load("en_core_web_sm")
doc=[]
doc = [nlp(news) for news in news_sample]

data_samples = []
# print(doc)
for i,token_doc in enumerate(doc):
  # print(token_doc)
  for token in token_doc:
    data_samples.append([token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop])
    # print(data_sample)

df = pd.DataFrame(data_samples, columns=["token.text", "token.lemma_", "token.pos_", "token.tag_", "token.dep_","token.shape_", "token.is_alpha", "token.is_stop"])

# # Display the DataFrame
print(df)
```

```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')

label_news= []
label1=[]
for data in news_sample:
  result = classifier(data)
  label = result[0]['label']
  label1.append(label)
  label_news.append({'Headline': data, 'Sentiment': label})
df = pd.DataFrame(label_news)
print(df)
```

```python

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_selection import chi2

df = pd.DataFrame({'News': news_sample, 'Label': label1})



count_vectorizer = CountVectorizer()
count_vect = count_vectorizer.fit_transform(df['News'])
# finding document freq
doc_freq = count_vect.sum(axis=0).A1


# Calculating TF-IDF scores
tfidf_vectorizer = TfidfVectorizer()
count_tfidf = tfidf_vectorizer.fit_transform(df['News'])
tfidf_scores = count_tfidf.mean(axis=0).A1

hi2_scores, p_values = chi2(count_vect, label1)


# here i am converting all the features into dataframes
df_features = pd.DataFrame(doc_freq, index=count_vectorizer.get_feature_names_out(), columns=['DF'])
df_tfidf = pd.DataFrame(tfidf_scores, index=tfidf_vectorizer.get_feature_names_out(), columns=['TF-IDF'])
df_chi2 = pd.DataFrame(hi2_scores, index=count_vectorizer.get_feature_names_out(), columns=['CHI'])

print(df_features)
print(df_tfidf)
print(df_tfidf)

# here i am joining them
feature_importance=[]
feature_importance = df_features.join(df_tfidf).join(df_chi2)

# ranking the features based on their importance by using "rank"
feature_importance['Rank by DF'] = feature_importance['DF'].rank(ascending=False)
feature_importance['Rank by TF-IDF'] = feature_importance['TF-IDF'].rank(ascending=False)
feature_importance['Rank by CHI'] = feature_importance['CHI'].rank(ascending=False)

# Displaying  the results
print(feature_importance)


# important features are DF, TF-IDF, Label encoder and chi extracted above,
#ranking the features based on their importance:
# 1)DF
# 2)tf-idf
# 3)chi
```

```python
def get_bert_tokenize(text):
  encoding = tockenizer.encode_plus(
    text,
    return_token_type_ids=False,
    padding="max_length",
    truncation=True,
    return_attention_mask=True,
    return_tensors='pt')
  return encoding

word_embeddings=[]


def get_bert_embeddings(text):
    # Tokenize and encode the text
    inputs = get_bert_tokenize(text)

    with torch.no_grad():
        outputs = model(**inputs)

    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

for news in news_sample:
  word_embeddings.append(get_bert_embeddings(news))
```
