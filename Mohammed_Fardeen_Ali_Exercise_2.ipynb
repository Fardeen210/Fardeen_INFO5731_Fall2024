## Question 2 (10 Points)
Write Python code to collect a dataset of 1000 samples related to the question discussed in Question 1.

```python
# write your answer here
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://github.com/marketplace?page=2&type=actions"
all_items = []
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:123.0) Gecko/20100101 Firefox/123.0",
    "Accept": "text/html",
    "Connection": "keep-alive"
}
for i in range(1,65):
    url = f"https://github.com/marketplace?page={i}&type=actions"
    response = requests.get(url, headers=headers)
    response_result = BeautifulSoup(response.content, 'html.parser')
    main = response_result.find_all(attrs={"data-testid": "non-featured-item"})
    # Appending the found items to the list
    all_items.extend(main)
```

```python
count=0
feauture_Name=[]
feauture_Description=[]

for i in all_items:
    # Find all <a> tags with the specified class inside the current element
    heading = i.find_all("a", class_="marketplace-module__marketplace-item-link--lsn3d line-clamp-1")
    heading_description= i.findAll("p")
    # Loop through the found <a> tags and print their content
    for link in heading:
        heading_name= link.get_text(strip=True)
        feauture_Name.append(heading_name)
    for desc in heading_description:
        description=desc.get_text(strip=True)
        feauture_Description.append(description)

df=pd.DataFrame(({
        'Heading': feauture_Name,
        'Description': feauture_Description
    }))
df.to_csv('github_data.csv', index=False)

for i in range(len(feauture_Name)):
    print(f"Feature {i}")
    print(f"Heading: {feauture_Name[i]}")
    print(f"Description: {feauture_Description[i]}")
    print('\n')
```

```python
import requests
from bs4 import BeautifulSoup
import time

headers = {
    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:123.0) Gecko/20100101 Firefox/123.0",
    "Accept": "text/html",
    "Connection": "keep-alive"
}
results=[]
for i in range(1,25):
  url = f"https://dl.acm.org/action/doSearch?AllField=xyz&EpubDate=%5B20190913+TO+202409132359%5D&queryID=12%2F7584138766&startPage={i}&pageSize=50"
  time.sleep(10)
  response = requests.get(url, headers=headers)
  response_result = BeautifulSoup(response.text, 'html.parser')
  main = response_result.findAll("li",class_="search__item issue-item-container")
  results.extend(main)
```

```python
titles=[]
journals=[]
year_published=[]
authors=[]
abstract=[]
# print(results)
for i in results:
  title=i.find("h5",class_="issue-item__title")
  titles.append(title.get_text(strip=True))

  journal_name=i.find("span",class_="epub-section__title")
  if journal_name:
    journals.append(journal_name.get_text(strip=True))
  else:
    journals.append("N/A")

  year=i.find("div",class_="bookPubDate")
  year_published.append(year.get_text(strip=True))

  author_names = i.find_all("span", class_="hlFld-ContribAuthor")
  authors.append([author.get_text(strip=True) for author in author_names])

  abstract_tag = i.find("div", class_="issue-item__abstract")
  abstract.append(abstract_tag.get_text(strip=True))

df = pd.DataFrame({
    'Title': titles,
    'Journal': journals,
    'Year Published': year_published,
    'Authors': authors,
    'Abstract': abstract
})
# Exporting the DataFrame to a CSV file
df.to_csv('articles_data.csv', index=False)

def printing_articles(num1,num2):
  for i in range(num1,num2):
    if not titles[i] or not journals[i] or not year_published[i] or not abstract[i]:
      continue
    print(f"Article {i}")
    print(f"Title: {titles[i]}")
    print(f"Journal: {journals[i]}")
    print(f"Year Published: {year_published[i]}")
    print(f"Authors: {', '.join(authors[i])}")
    print(f"Abstract: {abstract[i]}")
    print('\n')

printing_articles(0,501)
printing_articles(500,999)


# print(df.head(10))
```
